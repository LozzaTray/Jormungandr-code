```python
from model.graph_mcmc import Graph_MCMC
```

```python
graph = Graph_MCMC([])
graph.read_from_file("generated.gml")
```

```python
graph.partition(B_min=2, B_max=5)
```

```python
classifier = graph.sample_classifier_mcmc(2500, verbose=True)
```

```python
import numpy as np
```

```python
av_weight = np.zeros_like(classifier.weight_history["W1"][0])
square_weight = np.zeros_like(av_weight)

n = len(classifier.weight_history["W1"])

for i in range(0, n):
    weight = classifier.weight_history["W1"][i]
    av_weight += weight
    square_weight += weight ** 2

av_weight = av_weight / n
square_weight = square_weight / n
print(av_weight)
print(square_weight)
```
